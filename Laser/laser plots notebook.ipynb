```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load data from CSV
data = pd.read_csv("C:\\Users\\jasonbane\\Desktop\\nexo_code\\nexo\\Laser\\3 khz prf current v power.csv")

# Extract columns for plotting
x = data.iloc[:, 0]  # First column as x-axis
y = data.iloc[:, 1]  # Second column as y-axis

# Create a figure with a specified size
plt.figure(figsize=(10, 6))  # Adjust width and height as needed

# Plot the smoothed data and scatter plot
plt.scatter(x, y, color="red", marker="o", s=9, label="Pulse Data")

# Add labels, title, and grid
plt.xlabel("Current (A)")
plt.ylabel("Power (W)")
plt.title("Current (A) vs. Power (W)")
plt.grid(True)

# Customize ticks
x_ticks = np.arange(start=np.floor(min(x)), stop=np.ceil(max(x)) + 1, step=1)
plt.xticks(x_ticks)
y_ticks = np.arange(start=round(min(y)), stop=np.ceil(max(y)) + 1, step=2)
plt.yticks(y_ticks)

# Show plot
plt.show()
```

```python
import os

# Define file paths for pulse width and power data
pulse_files = [
    ("C:\\Users\\jasonbane\\Desktop\\nexo_code\\nexo\\Laser\\300 hz prf current v pulse width.csv", "Pulse Width - 300 Hz"),
    ("C:\\Users\\jasonbane\\Desktop\\nexo_code\\nexo\\Laser\\1 khz prf current v pulse width.csv", "Pulse Width - 1 kHz"),
    ("C:\\Users\\jasonbane\\Desktop\\nexo_code\\nexo\\Laser\\3 khz prf current v pulse width.csv", "Pulse Width - 3 kHz"),
]

power_files = [
    ("C:\\Users\\jasonbane\\Desktop\\nexo_code\\nexo\\Laser\\300 hz prf current v power.csv", "Power - 300 Hz"),
    ("C:\\Users\\jasonbane\\Desktop\\nexo_code\\nexo\\Laser\\1 khz prf current v power.csv", "Power - 1 kHz"),
    ("C:\\Users\\jasonbane\\Desktop\\nexo_code\\nexo\\Laser\\3 khz prf current v power.csv", "Power - 3 kHz"),
]

# Define colors and markers for each dataset
power_colors = ["red", "green", "blue"]
pulse_colors = ["pink", "lightgreen", "lightblue"]
markers = ["o", "s", "D"]

miny = float("inf")

# Output directory
output_dir = "C:\\Users\\jasonbane\\Desktop\\nexo_code\\nexo\\Laser\\plots\\"
os.makedirs(output_dir, exist_ok=True)

# Create a figure with a specified size
fig, ax1 = plt.subplots(figsize=(12, 7))

# Plot all power datasets
for i, (file_path, label) in enumerate(power_files):
    data = pd.read_csv(file_path)
    x = data.iloc[:, 0]
    y = data.iloc[:, 1]

    miny = min(y) if min(y) < miny else miny
    
    # Plot power data on primary y-axis
    ax1.scatter(x, y, color=power_colors[i], marker=markers[i], s=15, label=label)

# Create a secondary y-axis for pulse width data
ax2 = ax1.twinx()

# Plot all pulse width datasets
for i, (file_path, label) in enumerate(pulse_files):
    data = pd.read_csv(file_path)
    x = data.iloc[:, 0]
    y = data.iloc[:, 1]
    miny = min(y) if min(y) < miny else miny
    
    # Plot pulse width data on secondary y-axis
    ax2.scatter(x, y, color=pulse_colors[i], marker=markers[i], edgecolor='k', s=15, label=label)

# Set labels and titles
ax1.set_xlabel("Current (A)")
ax1.set_ylabel("Power (W)", color="black")
ax2.set_ylabel("Pulse Width (ns)", color="black")

# Customize ticks for x-axis and y-axes
x_ticks = np.arange(start=np.floor(min(x)), stop=np.ceil(max(x)) + 2, step=1)
ax1.set_xticks(x_ticks)
ax1.set_xticklabels(x_ticks, rotation=45, ha='right')

y_ticks_power = np.arange(start=round(miny), stop = 50, step=3)
ax1.set_yticks(y_ticks_power)

y_ticks_pulse = np.arange(start=round(miny), stop=np.ceil(max(y)) + 20, step=20)
ax2.set_yticks(y_ticks_pulse)

# Add legends for both y-axes
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()


# Add title and grid
plt.title("Current (A) vs. Power and Pulse Width for 300 - 3000 Hz")
ax1.grid(True, which='both', axis='both')

# Adjust margins to remove extra space and improve layout
plt.subplots_adjust(top=0.9, bottom=0.15, left=0.1, right=0.85)

# font = {'family' : 'normal',
#         'weight' : 'bold',
#         'size'   : 14}

# plt.rc('font', **font)
ax1.legend(lines1 + lines2, labels1 + labels2, loc="upper left", fontsize="small")
# Save plot
plot_filename = os.path.join(output_dir, "combined_plot.png")
ax1.set_rasterized(True)
ax2.set_rasterized(True)
plt.axvspan(26.0, 30.0, color="yellow", alpha=0.3, rasterized=True)
plt.savefig(plot_filename)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

pulse_files = {
    "300 Hz PRF": "C:\\Users\\jasonbane\\Desktop\\nexo_code\\nexo\\Laser\\300 hz prf current v pulse width.csv",
    "1 kHz PRF": "C:\\Users\\jasonbane\\Desktop\\nexo_code\\nexo\\Laser\\1 khz prf current v pulse width.csv",
    "3 kHz PRF": "C:\\Users\\jasonbane\\Desktop\\nexo_code\\nexo\\Laser\\3 khz prf current v pulse width.csv",
}

power_files = {
    "300 Hz PRF": "C:\\Users\\jasonbane\\Desktop\\nexo_code\\nexo\\Laser\\300 hz prf current v power.csv",
    "1 kHz PRF": "C:\\Users\\jasonbane\\Desktop\\nexo_code\\nexo\\Laser\\1 khz prf current v power.csv",
    "3 kHz PRF": "C:\\Users\\jasonbane\\Desktop\\nexo_code\\nexo\\Laser\\3 khz prf current v power.csv",
}

# Output directory
output_dir = "C:\\Users\\jasonbane\\Desktop\\nexo_code\\nexo\\Laser\\plots\\"
os.makedirs(output_dir, exist_ok=True)

# Define colors for each dataset
colors = {
    "300 Hz PRF": "red",
    "1 kHz PRF": "green",
    "3 kHz PRF": "blue",
}

# Function to create and save plots for pulse width and power
def create_combined_plot(files, y_label, title, filename_suffix, output_dir):
    fig, ax = plt.subplots(figsize=(12, 7))
    
    # Plot each dataset
    miny = float("inf")
    for label, file_path in files.items():
        data = pd.read_csv(file_path)
        x = data.iloc[:, 0]
        y = data.iloc[:, 1]
        
        miny = min(y) if min(y) < miny else miny

        color = colors[label]
        ax.plot(x, y, "-.", color=color, label=label )
    
    ax.set_xlabel("Current (A)")
    ax.set_ylabel(y_label)
    ax.set_title(title)

    # Customize ticks and grid
    x_ticks = np.arange(start=round(min(x)), stop=round(max(x))+2, step=1)
    ax.set_xticks(x_ticks)
    ax.set_xticklabels(x_ticks, rotation=45, ha='right')
    
    if max(y) > 330:
        y_ticks = np.arange(start=round(miny), stop=round(max(y)), step=15)
    else:
        y_ticks = np.arange(start=round(miny), stop=round(max(y)), step=2)
    ax.set_yticks(y_ticks)
    # ax.set_ylim(bottom=miny)

    ax.grid(True, which='both', axis='both')
    ax.legend()
    # font = {'family' : 'normal',
    #     'weight' : 'bold',
    #     'size'   : 15}

    # plt.rc('font', **font)
    # plt.subplots_adjust(top=0.9, bottom=0.2, left=0.1, right=0.85)

    # Save plot
    plot_filename = os.path.join(output_dir, f"{filename_suffix}.png")
    ax.set_rasterized(True)
    plt.axvspan(26.0, 30.0, color="yellow", alpha=0.3, rasterized=True)
    plt.savefig(plot_filename)
    plt.show()

# Create combined plots for pulse width and power
create_combined_plot(
    pulse_files,
    y_label="Pulse Width (ns)",
    title="Pulse Width vs. Current",
    filename_suffix="pulse_width_combined",
    output_dir=output_dir
)

create_combined_plot(
    power_files,
    y_label="Power (W)",
    title="Power vs. Current",
    filename_suffix="power_combined",
    output_dir=output_dir
)
```
