```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
DF = pd.read_table("Boiling_point.txt",skiprows=[0,1],names=["Pressure","Temp","Error"])
DF
```

```python
fig = plt.figure()

ax = fig.add_subplot(111)


ax.errorbar(x=DF["Temp"],xerr=DF["Error"],y=DF["Pressure"]*0.145038,label="Boiling")

ax.grid(True)
ax.set_ylabel("Pressure(psia)")
ax.set_xlabel("Temp(K)")


ax.set_yscale("log")
```
